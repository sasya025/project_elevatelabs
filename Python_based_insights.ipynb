```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
# Load data
df = pd.read_csv('retail_transaction_data.csv')

# Correlation
correlation = df[['Inventory_Days', 'Profit']].corr()
print("ðŸ”¹ Correlation Matrix:\n", correlation)

# Visualization
sns.regplot(x='Inventory_Days', y='Profit', data=df)
plt.title('Correlation Between Inventory Days and Profit')
plt.tight_layout()
plt.show()
```
